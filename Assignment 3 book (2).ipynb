```python
Question 2
from sklearn import datasets
iris = datasets.load_iris()
x=iris.data
y=iris.target
print (x.shape)
print(y)
```

```python
#transform x to 2d
from sklearn.decomposition import PCA
pca=PCA(n_components = 2)
pca.fit(x)
x_2D=pca.transform(x)
print (x_2D)
```

```python
from sklearn.model_selection import train_test_split
Xtrain, Xtest, ytrain, ytest=train_test_split(x_2D, y, test_size=0.4, random_state=1)
```

```python
#apply KNN model
from sklearn.neighbors import KNeighborsClassifier
classifier= KNeighborsClassifier(n_neighbors=5)
classifier.fit(Xtrain, ytrain)
y_predict_train=classifier.predict(Xtrain)
y_predict_test=classifier.predict(Xtest)
```

```python
from sklearn.metrics import accuracy_score
acc_train = accuracy_score(ytrain, y_predict_train)
acc_test = accuracy_score(ytest, y_predict_test)
print(acc_train)
print(acc_test)
```

```python
#classify the model boundry 
import matplotlib.pyplot as plt
from mlxtend.plotting import plot_decision_regions
fig=plt.figure(figsize =(15,6))
ax1=plt.subplot(1,2,1)
plot_decision_regions(X=Xtrain, y=ytrain, clf=classifier, legend=2)
plt.text(3,-1.2,round(acc_train,5))

ax2=plt.subplot(1,2,1)
plot_decision_regions(X=Xtest, y=ytest, clf=classifier, legend=2)
plt.text(3,-1.2,round(acc_test,5))
```
